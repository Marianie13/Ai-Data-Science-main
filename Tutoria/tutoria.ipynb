```python
import pandas as pd 

#cargar el archivo csv
df = pd.read_csv("hotel_bookings.csv")

#Mostrar ñas primeras filas del dataframe
df
```

```python
#Informacion general del dataframe
print(df.info())
```

```python
#Mostrar todas las columnas del dataframe
for col in df.columns:
    print(col)
```

```python
#Verificar si hay valores nulos en el dataframe
print(df.isnull().sum())
```

```python
#Remplazar valores nulos en 'children' con 0
df['children'].fillna(0, inplace=True)
```

```python
#Verificar si hay valores nulos en el dataframe
print(df.isnull().sum())
```

```python
#Rellenar valores nulos en 'country' con 'Unknown'
df['country'].fillna('Unknown', inplace=True)
```

```python
#Verificar si hay valores nulos en el dataframe
print(df.isnull().sum())
```

```python
#Rellenar valores nulos en 'agent' y 'company' con 0
df['agent'].fillna(0, inplace=True)
df['company'].fillna(0, inplace=True)
```

```python
#Verificar si hay valores nulos en el dataframe
print(df.isnull().sum())
```

```python
#Guardar el dataframe Limpio en un nuevo archivo csv
df.to_csv("hotel_bookings_clean.csv", index=False)
```

```python
#Contar reservar por tipo de hotel
hotel_counts = df['hotel'].value_counts()

print(hotel_counts)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Graficar la cantidad de reservas por tipo de hotel
sns.barplot(x=hotel_counts.index, y=hotel_counts.values)
plt.title('Cantidad de Reservas por Tipo de Hotel')
plt.xlabel('Tipo de Hotel') 
plt.ylabel('Numero de Reservas')
plt.show()
```

```python
#calcular la taba de cancelacion
cancelation_rate = df['is_canceled'].mean()

print(f"Tasa de cancelacion: {cancelation_rate:.2%}")
```

```python
#ingresos promedio por tipo de hotel
average_adr  = df.groupby('hotel')['adr'].mean()
print("Ingresos promedio por tipo de hotel:")
for hotel, adr in average_adr.items():
    print(f"{hotel}: {adr:.2f} $")
```

```python
#Tasa de cancelacion por tipo de cliente
cancellation_by_customer_type = df.groupby('customer_type')['is_canceled'].mean()
print("Tasa de cancelacion por tipo de cliente:")
for customer_type, rate in cancellation_by_customer_type.items():
    print(f"{customer_type}: {rate:.2%}")
```

```python
#Crear una columana de fechas
df['arrival_date'] = pd.to_datetime(df['arrival_date_year'].astype(str) + '-' + df['arrival_date_month'] + '-' + df['arrival_date_day_of_month'].astype(str))

#Contar reservas por mes
monthly_counts = df['arrival_date'].dt.to_period('M').value_counts().sort_index()

print(monthly_counts)
```

```python
monthly_counts.plot(kind='line')
plt.title('Reservas por Mes')
plt.xlabel('Mes')
plt.ylabel('Numero de Reservas')
plt.show()
```

```python
#Grafica de barras
monthly_counts.plot(kind='bar', color='skyblue')
plt.title('Reservas por Mes')
plt.xlabel('Mes')
plt.ylabel('Numero de Reservas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
